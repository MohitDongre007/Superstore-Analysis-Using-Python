```python
#importing neccessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import datetime as datetime 
```

```python
#reading data

df = pd.read_csv('/kaggle/input/superstore-dataset-final/Sample - Superstore.csv', encoding='Latin-1')
```

```python
#changing data type

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df['Order Month'] = df['Order Date'].dt.month 
df['Order Year'] = df['Order Date'].dt.year 
df['Order Day of Week'] = df['Order Date'].dt.dayofweek 
```

```python
df.info()
```

# Let’s have a look at the Monthly Sales:

```python
sales_by_month = df.groupby('Order Month')['Sales'].sum().reset_index()

fig = px.line(sales_by_month, x='Order Month', y='Sales')

fig.update_layout(title_text='Monthly Sales Analysis' , title_font=dict(size=24))

fig.show()
```

# Let’s have a look at the sales by Category:

```python
sales_by_category = df.groupby('Category')['Sales'].sum().reset_index()

fig = px.pie(sales_by_category, names='Category', values='Sales', hole=0.4)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Sales Analysis by Category', title_font=dict(size=24))

fig.show()
```

#  Let’s have a look at the sales by Sub-Category:

```python
sales_by_subcategory = df.groupby('Sub-Category')['Sales'].sum().reset_index()

fig = px.bar(sales_by_subcategory, x='Sub-Category', y='Sales')

fig.update_layout(title="Sales Analysis by Sub-Category", title_font= dict(size=24))

fig.show()
```

# Let’s have a look at the Monthly Profits:

```python
monthly_profit = df.groupby('Order Month')['Profit'].sum().reset_index()

fig = px.line(monthly_profit, x='Order Month', y='Profit')

fig.update_layout(title="Monthly Profit Analysis", title_font=(dict(size=24)))

fig.show()
```

# Let’s have a look at the profit by Category:

```python
profit_by_category = df.groupby('Category')['Profit'].sum().reset_index()

fig = px.pie(profit_by_category, names='Category', values='Profit', hole=0.4)

fig.update_layout(title="Profit Analysis by Category", title_font=dict(size=24))
fig.update_traces(textposition='inside', textinfo= 'percent+label')

fig.show()
```

# Let’s have a look at the profit by Sub-Category:

```python
profit_by_subcategory = df.groupby('Sub-Category')['Profit'].sum().reset_index()

fig = px.bar(profit_by_subcategory, x='Sub-Category', y='Profit')

fig.update_layout(title="Profit Analysis by Sub-Category", title_font=(dict(size=24)))

fig.show()
```

# Let’s have a look at the sales and profit analysis by Customer Segments:

```python
sales_profit_by_segment = df.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_by_segment
```

```python
fig = px.bar(sales_profit_by_segment, x="Segment", y=['Sales','Profit'], barmode='group')

fig.update_layout(title="Sales and Profit Analysis by Customer Segment", title_font=dict(size=24), xaxis_title="Customer Segment", yaxis_title="Amount")

fig.show()
```

**So the store has higher profits from the product sales for consumers, but the profit from corporate product sales is better in the sales-to-profit ratio. Let’s have a look at it to validate our findings:**

```python
sales_profit_by_segment = df.groupby('Segment').agg({'Sales':'sum', 'Profit':'sum'}).reset_index()
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment['Sales']/sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment','Sales_to_Profit_Ratio']])
```
